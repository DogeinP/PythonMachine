```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib
import matplotlib.font_manager as fm
path = 'C:/Users/user/ipyn/NanumFontSetup_TTF_ALL/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=path, size=10)
fontprop_title = fm.FontProperties(fname=path, size=30)

%matplotlib inline
```

```python
popular = pd.read_csv('report.txt', sep = "\t", engine='python', encoding = "UTF-8")
```

```python
popular.head()
```

```python
t = list(set(popular['기간']))
t.remove('기간')
t.sort()
```

```python
title = {'세대':2, '인구':3, '한국인 인구':6, '외국인 인구':9, '세대 당 인구':12, '고령자':13}
```

```python
for g in range(len(t)):
    df = popular.loc[popular.iloc[:,0]==t[g],:][1:]
    plt.subplot(len(title),1,1)
    print(t[g])
    for i,j in title.items():

        plt.figure(figsize=(20,10))
        plt.bar(df.iloc[:,1],[float(i.replace(',','')) for i in df.iloc[:,j]],)
        plt.xticks(df.iloc[:,1],list(df.iloc[:,1]),fontproperties=fontprop)
        plt.title(t[g]+' '+i,fontproperties=fontprop_title)
        plt.axhline(y=round(np.mean([float(i.replace(',','')) for i in df.iloc[:,j]]),1), color='r', linewidth=1)
        plt.show()
```
